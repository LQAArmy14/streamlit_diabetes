# Tentukan Library yang Digunakan

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

# Load Dataset


```python
diabetes_dataset = pd.read_csv('drive/MyDrive/PROJECT_AI/diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape #menampilkan jumlah baris dan kolom
```

```python
diabetes_dataset['Outcome'].value_counts() #melihat berapa data yang terkena dan tidak terkena diabetes
```

```python
#memisahkan data dan label
#data adalah atribut lain
#label adalah outcome

X = diabetes_dataset.drop (columns='Outcome', axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

# Standarisasi Data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

# Memisahkan Data Training dan Data Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Membuat Data Latih Menggunakan Algoritma SVM

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

# Membuat Model Evaluasi untuk Mengukur Tingkat Akurasi

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Akurasi Data Training adalah = ', training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Akurasi data testing adalah = ', test_data_accuracy)
```

# Membuat Model Prediksi

```python
input_data = (6, 148, 72, 35, 0, 33.6, 0.627, 50)

input_data_as_numpy_array = np.array(input_data)

input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
    print('Pasien Tidak Terkena Penyakit Diabetes ')
else :
    print('Pasien Terkena Penyakit Diabetes')
```
